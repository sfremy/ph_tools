```python
def planck_approx(x1,x2):
    h=6.62607015*(10**-34)
    e=2.718281828459
    c=299792458
    k=1.380649*(10**-23)
    T=5000
    y1 = 2*h*c**2 / ((x1**5)*(e**(h*c/x1/k/T)-1))
    y2 = 2*h*c**2 / ((x2**5)*(e**(h*c/x2/k/T)-1))

    approx = (y1+y2)*(x2-x1)/2
    return approx

planck_approx(397.5*(10**-9),492.5*(10**-9))/(5.67*10**-8*5000**4/3.141592653589793238462643383279)
```
